# Bar graph of male vs female buyers participated in the sales

```python
dataset ['Buyer Gender'].value_counts()
```

```python
Gender = dataset['Buyer Gender'].value_counts().to_frame().reset_index()
plt.figure(figsize=(8,10))
plt.bar(Gender['index'],Gender['Buyer Gender'],color = ['red','blue'])
plt.title('Female vs Male')
plt.xlabel('Gender')
plt.ylabel('Count') 
```

Insight: From the bar plot we can see that there are more female buyers than male buyers
